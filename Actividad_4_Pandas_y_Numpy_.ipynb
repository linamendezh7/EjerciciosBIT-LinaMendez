1. Cuál es el promedio de consumo de cafeina en las mujeres menores de 50 años por nivel de estres?

```python
valores_femeninos = ["Female", "female"] # Añade todas las variaciones posibles
Casos_femeninos = proyecto[proyecto["Gender"].isin(valores_femeninos)]
Casos_femeninos = Casos_femeninos[Casos_femeninos["Age"]<50]
Casos_femeninos
```

```python
Consumo_cafeina = Casos_femeninos['Caffeine Intake (mg/day)']
Consumo_cafeina = np.array(Consumo_cafeina)
```

```python
q1 = np.percentile(Consumo_cafeina, 25)
q2 = np.percentile(Consumo_cafeina, 50)
q3 = np.percentile(Consumo_cafeina, 75)
iqr = q3 - q1

print(f"El primer cuartil (percentil 25) es igual a: {q1}")
print(f"El segundo cuartil (percentil 50) es igual a: {q2}")
print(f"El tercer cuartil (percentil 75) es igual a: {q3}")
print(f"El rango intercuartil es igual a: {iqr}")
```

```python
import matplotlib.pyplot as plt
plt.boxplot(Casos_femeninos['Caffeine Intake (mg/day)'])
plt.title('Consumo de cafeina en mujeres menores de 50 años')
plt.ylabel('Cafeina (mg/day)')
plt.grid()
plt.show()
```

```python
data_barras = pd.DataFrame(Casos_femeninos['Caffeine Intake (mg/day)'].value_counts())
data_barras
```

```python
Cafeina_promedio = Casos_femeninos.groupby('Stress Level (1-10)')['Caffeine Intake (mg/day)'].mean()
duracion_promedio = Cafeina_promedio.sort_values()
duracion_promedio
```
