Ejercicios

1. Lista ProductID, Name, ListPrice de SalesLT.Product
(SELECT)

```python
Ejercicio1 = pd.read_sql_query("""
    SELECT ProductID, Name,ListPrice
    FROM SalesLT.Product
    """, conn)
Ejercicio1
```

2. Devuelve productos con ListPrice > 1000 mostrando ProductNumber, Name, ListPrice.

```python
Ejercicio2 = pd.read_sql_query("""
    SELECT ProductNumber, Name,ListPrice
    FROM SalesLT.Product
    WHERE ListPrice >1000
    """, conn)
Ejercicio2
```

3. Muestra el TOP 5 productos más caros con ProductID, Name, ListPrice.

```python
N = 5

Ejercicio3 = pd.read_sql_query(f"""
    SELECT TOP ({N}) ProductID, ProductNumber, Name, ListPrice
    FROM SalesLT.Product
    ORDER BY ListPrice DESC
    """, conn)
Ejercicio3
```

4. Lista ProductID, p.Name AS ProductName, c.Name AS CategoryName para productos con categoría
(INNER JOIN).

```python

Ejercicio4 = pd.read_sql_query(f"""
    SELECT
        p.ProductID,
        p.Name AS ProductName,
        c.Name AS CategoryName
    FROM SalesLT.Product AS p
    INNER JOIN SalesLT.ProductCategory AS c ON p.ProductCategoryID = c.ProductCategoryID;
    """, conn)
Ejercicio4
```

5. Obtén pedidos del año 2008 en adelante (o el año que aplique en tu base) con SalesOrderID, OrderDate, CustomerID, TotalDue y el nombre del cliente (FirstName + ' ' + LastName).
(JOIN + WHERE)

```python
Ejercicio5 = pd.read_sql_query(f"""
    SELECT
        c.CustomerID,
        c.FirstName,
        c.LastName,
        p.SalesOrderID,
        p.OrderDate,
        p.TotalDue
    FROM  SalesLT.Customer AS c
    INNER JOIN SalesLT.SalesOrderHeader AS p ON p.CustomerID = c.CustomerID
    WHERE p.OrderDate >= '2008-01-01'
    """, conn)
Ejercicio5
```

6. Para cada ProductID, suma la cantidad vendida (SUM(OrderQty)). Devuelve solo productos con ventas totales > 1000, ordenados de manera descendente.
Usa GROUP BY, HAVING y ORDER BY

```python
Ejercicio6 = pd.read_sql_query(f"""
    SELECT
        ProductID,
        ROUND(SUM(LineTotal),1) AS Vta_Total,
        SUM(OrderQty) AS Cantidad
    FROM  SalesLT.SalesOrderDetail
    GROUP BY ProductID
    HAVING SUM(LineTotal) > 1000
    ORDER BY Vta_Total DESC
    """, conn)
Ejercicio6
```

7. Productos sin ventas: lista productos que no aparecen en SalesLT.SalesOrderDetail.
Usa LEFT JOIN + WHERE

```python
Ejercicio7 = pd.read_sql_query(f"""
    SELECT
        p.ProductID,
        c.name
    FROM  SalesLT.Product AS c
    LEFT JOIN SalesLT.SalesOrderDetail AS p ON p.ProductID = c.ProductID
    WHERE p.ProductID IS NULL
    """, conn)
Ejercicio7
```

8. Clientes por país: total de clientes por CountryRegion (usa Address + CustomerAddress).
Podrias necesitar: COUNT(DISTINCT CustomerID).

```python
Ejercicio8 = pd.read_sql_query(f"""
    SELECT
        p.StateProvince,
        COUNT(DISTINCT c.CustomerID) AS TotalClientes
    FROM SalesLT.CustomerAddress AS c
    LEFT JOIN
        SalesLT.Address AS p ON p.AddressID = c.AddressID
    GROUP BY p.StateProvince
    ORDER BY TotalClientes DESC
    """, conn)
Ejercicio8
```

9. Ticket promedio por año: AVG(TotalDue) por YEAR(OrderDate).

```python
Ejercicio9 = pd.read_sql_query(f"""
    SELECT
        p.OrderDate,
        ROUND(AVG(p.TotalDue),1) AS Ticket_promedio
    FROM  SalesLT.SalesOrderDetail AS c
    INNER JOIN SalesLT.SalesOrderHeader AS p ON p.SalesOrderID = c.SalesOrderID
    GROUP BY p.OrderDate
    """, conn)
Ejercicio9
```
