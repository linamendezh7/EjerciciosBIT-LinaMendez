Actividad 1 - Estructuras y tipos de datos
Objetivo: Administrar un inventario de una tienda

Añade un nuevo producto al inventario con las siguientes caracteristicas:
{"producto": "abrigo", "precio": 50000, "stock": "2"}


```python
inventario = [
    {"producto": "camisa", "precio": 25900, "stock": "11"},
    {"producto": "pantalón", "precio": 39900, "stock": "23"},
]
inventario
```

```python
productoadicional = {
    "producto": "abrigo",
    "precio": 50000,
    "stock": "2"}

inventario.append(productoadicional)
inventario
```

Convierte los valores del stock en numeros enteros

```python
for inv in inventario:
    inv ['stock'] = int(inv.get("stock"))
inventario
```

 Actualiza el precio de cada producto sumando 10000 al precio original Imprime cada producto del inventario y sus caracteristicas con el siguiente formato:
 print(f"Hay {stock} unidades del producto {producto}. Su precio por unidad es de {precio}")

```python
for price in inventario:
    price ['precio'] = price.get("precio")+10000
inventario
```

```python
for tabla in inventario:
    pro = tabla["producto"]
    pre = tabla["precio"]
    sto = tabla["stock"]
    print(f"Hay {sto} unidades del producto {pro}. Su precio por unidad es de {pre}")
```
